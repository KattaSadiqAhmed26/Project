```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\hyper\OneDrive\Desktop\Internship\Project1\retail_sales.csv')
```

```python
df.shape
```

```python
# Correlation Between Inventory_Days and Profitability.
correlation = df['Inventory_Days'].corr(df['Profit'])
print(f"Correlation between Inventory_Days and Profit : {correlation:.2f}")
```

```python
# Scatter plot Inventory Days vs Profit.
sns.scatterplot(data=df , x='Inventory_Days', y='Profit', hue='Category')
plt.title('Inventory Days VS Profit')
plt.show()
```

```python
# Profit Margin by Category.
df['Profit_Margin'] = df['Profit'] / df['Sales']
sns.boxplot(data=df , x='Category', y='Profit_Margin')
plt.title('Profit Margin Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```
